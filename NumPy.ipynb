## Numpy
Numerical Python.

It is a library consisting of multidimensional array objects and a collection of routines for processing of array.
It also has functions for working in domain of linear algebra, fourier transform, and matrices.



```python
import numpy as np
arr = np.array([1, 2, 3, 4, 5]) #index from 0
print(arr)
print(type(arr))

#The array object in NumPy is called ndarray
```

Every item in an ndarray takes the same size of block in the memory. Each element in ndarray is an object of data-type object (called dtype). Any item extracted from ndarray object (by slicing) is represented by a Python object of one of array scalar types.



```python
import numpy as np

a = np.array([1, 2, 3])
print('a =', a)
print('type(a) =', type(a))
print('a.dtype =', a.dtype)
print('a.ndim =', a.ndim)
print('a.size =', a.size)
print('a.shape =', a.shape)
```

```python
b = np.array([[1.3, 2.4], [0.3, 4.1]])
print('b =\n', b)
print('type(b) =', type(b))
print('b.dtype =', b.dtype)
print('b.ndim =', b.ndim)
print('b.size =', b.size)
print('b.shape =', b.shape)
```

- numpy.empty: It creates an uninitialized array of specified shape and dtype.
- numpy.zeros: Returns a new array of specified size, filled with zeros.
- numpy.ones: Returns a new array of specified size, filled with ones.

```python
x = np.empty([3,2], dtype = int) 
print(x)

y = np.zeros(5)
print(y)

z = np.zeros((5,), dtype=int)
print(z)

f = np.ones([2,2], dtype = int) 
print(f)
```

### Arithmetic Operators

- np.sum(a): sum of all elements of a matrix
- np.sum(a, axis=0): Sum of elements of a matrix column-wise
- np.sum(a,axis=1): Sum of elements of a matrix row-wise 

### Matrix Operations
- np.add(): addition of two matrices
- np.subtract(): Subtraction of two matrices
- np.multiply(): Multiplication of two matrices
- np.divide(): Division of two matrices
- np.dot(): matrix multiplication

```python
a=[[0,1],[0,5]]
print(a)
print(np.sum(a))
print(np.sum(a,axis=0))
print(np.sum(a,axis=1))
```

```python
a=np.array([[0,1,2],[2,3,4]])
b=np.array([4,5,6])
print(np.add(a,b))
print(np.subtract(a,b))
print(np.multiply(a,b))
print(np.divide(a,b))
print(np.dot(a,b))
```

The numpy.average() function computes the weighted average of elements in an array according to their respective weight given in another array.

Considering an array [1,2,3,4] and corresponding weights [4,3,2,1], the weighted average is calculated by adding the product of the corresponding elements and dividing the sum by the sum of weights.

Weighted average = (1\*4+2\*3+3\*2+4\*1)/(4+3+2+1)


```python
a = np.array([1,2,3,4]) 
print(np.average(a))
wts = np.array([4,3,2,1]) 
print(np.average(a,weights=wts))
```
